```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mlp_df = 'mlp_results.csv'
kan_df = 'kan_results.csv'
```

```python
mlp_df = pd.read_csv(mlp_df)
kan_df = pd.read_csv(kan_df)
```

```python
print(mlp_df.columns)
mlp_df.head()
```

```python
plt.style.use('ggplot')
plt.subplots(figsize = (15,10))
epochs = mlp_df['epoch']
plt.plot(epochs, mlp_df[' test_acc'], color = 'blue', label = 'MLP Test acc', marker = 'o')
plt.plot(epochs, kan_df[' test_acc'], color = 'red', label = 'KAN Test acc', marker = 'o')
plt.plot(epochs, mlp_df[' train_acc'], linestyle = '--', color = 'blue', label = 'MLP Training acc', marker = 'o')
plt.plot(epochs, kan_df[' train_acc'], linestyle = '--', color = 'red', label = 'KAN Training acc', marker = 'o')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.title("MNIST Accuracy using HSIC Bottleneck Training")
```
