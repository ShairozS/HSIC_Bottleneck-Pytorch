```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mlp_df = 'mlp_results.csv'
kan_df = 'kan_results.csv'
```

```python
mlp_df = pd.read_csv(mlp_df)
kan_df = pd.read_csv(kan_df)
```

```python
print(mlp_df.columns)
mlp_df.head()
```

```python
plt.style.use('ggplot')
plt.subplots(figsize = (15,10))
epochs = mlp_df['epoch']
plt.plot(epochs, mlp_df[' test_acc'], color = 'blue', label = 'MLP Test acc', marker = 'o')
plt.plot(epochs, kan_df[' test_acc'], color = 'red', label = 'KAN Test acc', marker = 'o')
plt.plot(epochs, mlp_df[' train_acc'], linestyle = '--', color = 'blue', label = 'MLP Training acc', marker = 'o')
plt.plot(epochs, kan_df[' train_acc'], linestyle = '--', color = 'red', label = 'KAN Training acc', marker = 'o')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.title("MNIST Accuracy using HSIC Bottleneck Training")
```

```python
import numpy as np

plt.style.use('ggplot')
plt.subplots(figsize = (15,10))
epochs = mlp_df['epoch']
total_mlp = np.sum(mlp_df[' time']).round(2)
total_kan = np.sum(kan_df[' time']).round(2)

plt.plot(epochs, mlp_df[' time'], color = 'blue', label = 'MLP (total = ' + str(total_mlp) + " seconds)", marker = 'o')
plt.plot(epochs, kan_df[' time'], color = 'red', label = 'KAN (total = ' + str(total_kan) + " seconds)", marker = 'o')

plt.xlabel("Epochs")
plt.ylabel("Time (in seconds)")
plt.legend()
plt.title("MNIST Training Time using HSIC Bottleneck Training")
```
